```r
NRows = 100
NCols = 30
m1 <- matrix(runif(NCols*NRows,min = 1, max = 200),ncol=NCols)
m1
head(m1)
```

```r
m1[10:60,] <- NA
m1
#syntax on how to make specified values null
```

```r
missing_row_val <- sum(rowSums(is.na(m1))>0)
missing_row_val
```

```r
m1 <- apply(m1,2,function(x) ifelse(is.na(x), mean(x,na.rm = TRUE),x))
m1
```

```r
correlation_matrix <- cor(m1)
correlation_matrix
```

```r
plot(m1)
```

```r
heatmap(correlation_matrix, col = colorRampPalette(c("blue", "white", "red"))(100), 
        main = "Pearson Correlation Heatmap")
```

```r
columns_to_keep <- colnames(correlation_matrix)[!(apply(abs(correlation_matrix) > 0.7, 2, any))]
columns_to_keep
```

```r
normalized_dataset <- apply(m1, 2, function(x) (x - min(x, na.rm = TRUE)) / (max(x, na.rm = TRUE) - min(x, na.rm = TRUE)) * 10)
normalized_dataset
```

```r
binary_dataset <- ifelse(normalized_dataset <= 0.5, 1, 0)
binary_dataset
```
