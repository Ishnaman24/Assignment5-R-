```r
# Set the random seed for reproducibility
set.seed(42)

# Generate random dataset
dataset <- matrix(runif(600*15, min = -100, max = 100), nrow = 600, ncol = 15)

# Convert the matrix to a data frame
dataset <- as.data.frame(dataset)

# Assign column names
colnames(dataset) <- paste0("Column", 1:15)

# (i) Plot scatter graph between Column 5 and Column 6
plot(dataset$Column5, dataset$Column6, xlab = "Column 5", ylab = "Column 6", main = "Scatter Plot")

#  Plot histogram of each column in a single graph
par(mfrow = c(3, 5))  # Set the layout of subplots to 3 rows and 5 columns
for (i in 1:15) {
  hist(dataset[, i], main = paste0("Histogram of ", colnames(dataset)[i]))
}

#  Plot the Box plot of each column in a single graph
boxplot(dataset, main = "Box Plot")
```
