```r
# Set the random seed for reproducibility
set.seed(42)

# Generate random dataset
dataset <- data.frame(matrix(0, nrow = 500, ncol = 10))
colnames(dataset) <- c("Col1", "Col2", "Col3", "Col4", "Col5", "Col6", "Col7", "Col8", "Col9", "Col10")

# Assign random values to columns 1-4 [-10, 10]
dataset[, 1:4] <- runif(500, min = -10, max = 10)

# Assign random values to columns 5-8 [10, 20]
dataset[, 5:8] <- runif(500, min = 10, max = 20)

# Assign random values to columns 9-10 [-100, 100]
dataset[, 9:10] <- runif(500, min = -100, max = 100)

# Apply K-Means clustering
kmeans_result <- kmeans(dataset, centers = 2, nstart = 25)

# Determine the optimal number of clusters using elbow method
wss <- sapply(1:10, function(k) kmeans(dataset, centers = k)$tot.withinss)
plot(1:10, wss, type = "b", xlab = "Number of Clusters", ylab = "Total Within Sum of Squares", main = "Elbow Method")

# Apply Hierarchical clustering
hclust_result <- hclust(dist(dataset))

# Plot the dendrogram for Hierarchical clustering
plot(hclust_result, main = "Dendrogram")
```
