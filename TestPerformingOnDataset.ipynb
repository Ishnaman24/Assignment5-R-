```r
# Set the random seed for reproducibility
set.seed(42)

# Generate random dataset
dataset <- matrix(runif(500*5, min = 5, max = 10), nrow = 500, ncol = 5)

# Convert the matrix to a data frame
dataset <- as.data.frame(dataset)

# Assign column names
colnames(dataset) <- paste0("Column", 1:5)

# Perform t-Test on each column
t_test_results <- lapply(dataset, function(x) t.test(x))

# Perform Wilcoxon Signed Rank Test on each column
wilcoxon_results <- lapply(dataset, function(x) wilcox.test(x))

#  Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4
col3 <- dataset$Column3
col4 <- dataset$Column4
two_sample_t_test_result <- t.test(col3, col4)
wilcoxon_rank_sum_test_result <- wilcox.test(col3, col4)


for (i in 1:5) {
  cat("Column", i, "t-Test Result:\n")
  print(t_test_results[[i]])
  cat("\nColumn", i, "Wilcoxon Signed Rank Test Result:\n")
  print(wilcoxon_results[[i]])
  cat("\n")
}

cat("Two Sample t-Test Result on Column 3 and Column 4:\n")
print(two_sample_t_test_result)
cat("\nWilcoxon Rank Sum Test Result on Column 3 and Column 4:\n")
print(wilcoxon_rank_sum_test_result)
```
